```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
data_path = "../Data/docs/4.papyrus_present_genes&compounds.csv"
df = pd.read_csv(data_path)
df
```

```python
# filter df symbol
symbol = 'ACVR2B'
symbol2 = 'ACVR1B'
df_target = df[df['Symbol'] == (symbol)]
df_suplement = df[df['Symbol'] == (symbol2)]
df_target
```

```python
# make list of SMILES column in df_targeted
smiles_list = df_target['SMILES'].tolist()
smiles_list2 = df_suplement['SMILES'].tolist()
print('ACVR2B molecules:', len(smiles_list))
print('ACVR molecules:', len(smiles_list2))
```

```python
total_smiles = len(smiles_list)  # Total SMILES in the list

# Filter rows where SMILES column matches any in the list
filtered_df = df[df['SMILES'].isin(smiles_list)]

# Remove duplicate (SMILES, Symbol) pairs to ensure each SMILES is counted only once per Symbol
filtered_df = filtered_df.drop_duplicates(subset=['SMILES', 'Symbol'])

# Exclude 'ACVR2B' from the analysis
filtered_df = filtered_df[filtered_df['Symbol'] != 'ACVR2B']

# Count occurrences of each Symbol
symbol_counts = filtered_df['Symbol'].value_counts()

# Convert counts to percentage of total SMILES list
symbol_percentage = (symbol_counts / total_smiles) * 100

# Select the top 15 symbols with the highest percentage
top_15_symbols = symbol_percentage.head(15)

# Plot the bar chart
plt.figure(figsize=(6, 5))
top_15_symbols.plot(kind='bar', color='grey', edgecolor='black', width=0.8)

# Beautify the plot
plt.xlabel("Genes")
plt.ylabel("Percentage (%)")
plt.title("Percentage of ACVR2B compounds not specific to ACVR2B")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 100)
plt.savefig("../Results/images/ACVR2B_compounds_percentage.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Generate RDKit molecule objects from the SMILES column
molecules = [Chem.MolFromSmiles(smiles) for smiles in df_short['SMILES']]

# Format the pchembl_value_Mean column to 3 decimal places
df_short['pchembl_value_Mean'] = df_short['pchembl_value_Mean'].round(3)

# Update legends with formatted values
legends = df_short['pchembl_value_Mean'].astype(str).tolist()

# Regenerate the grid image with updated legends
img = Draw.MolsToGridImage(molecules, molsPerRow=4, subImgSize=(400, 400), legends=legends)

with open("../Results/images/pharmacophore molecules.png", "wb") as png:
    png.write(img.data)
# Display the image
img
```

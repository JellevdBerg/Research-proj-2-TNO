```python
from PIL import Image
import pymol2
import os
import math
from IPython.display import display

df = pd.read_csv('../Data/docs/9.noisy_poses.csv')

img_paths = []
for cluster_id in df['cluster'].unique():
    with pymol2.PyMOL() as pymol:
        cmd = pymol.cmd
        cluster_df = df[df['cluster'] == cluster_id]

        for name in cluster_df['name']:
            sdf_path = f"../Data/other/SDF poses/{name}.sdf"
            if os.path.exists(sdf_path):
                cmd.load(sdf_path, name)
                cmd.show("sticks", name)

        cmd.bg_color('white')
        cmd.set('specular', 0)
        cmd.set('ray_shadows', 0)
        cmd.orient('All')
        cmd.ray()

        img_path = f"../Results/images/noisy_{cluster_id}.png"
        cmd.png(img_path, width=1000, height=1000)
        img_paths.append(img_path)

# Load images
images = [Image.open(p) for p in img_paths]
cols = math.ceil(len(images) / 3)
thumb_w, thumb_h = images[0].size

# Create blank canvas
combined = Image.new('RGB', (cols * thumb_w, 3 * thumb_h), 'white')

# Paste images row by row
for i, img in enumerate(images):
    x = (i % cols) * thumb_w
    y = (i // cols) * thumb_h
    combined.paste(img, (x, y))

combined_path = "../Results/images/all_noisy_clusters.png"
combined.save(combined_path)
```
